```python
import seaborn as sns
import pandas as pd
```

```python
p = pd.read_csv("preds_phenotypes_rs_5cl.csv").iloc[:, 1:]
p.iloc[-3:, 1] = 1250
xx = pd.melt(p, id_vars=["Run", "Weight"], value_vars=p.columns[2:], value_name="AUC", var_name="Phenotypes")
sns.lineplot(data=xx, x="Weight", y="AUC", hue="Phenotypes")
```
