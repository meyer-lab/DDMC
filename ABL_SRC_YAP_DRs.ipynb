```python
d = pd.read_csv("msresist/data/Validations/Luminex/RTK_luminex.csv")
d = d[d["Type"] == "Dil AB"]
d["Line"] = [s.split(" ")[0] for s in d["Sample"]]
d["Treatment"] = [s.split(" ")[1] for s in d["Sample"]]
d = d.drop(["Sample", "Unnamed: 4"], axis=1).iloc[1:, :]
data = pd.melt(d, id_vars=["Type", "Line", "Treatment"], value_vars=["P-AXL", "p-EGFR"], var_name="Protein", value_name="p-signal")

_, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.barplot(data=data[data["Protein"] == "P-AXL"], x="Treatment", y="p-signal", hue="Line", ax=ax[0])
ax[0].set_title("p-AXL Y702")
sns.barplot(data=data[data["Protein"] == "p-EGFR"], x="Treatment", y="p-signal", hue="Line", ax=ax[1])
ax[1].set_title("p-EGFR lookup site")
```
