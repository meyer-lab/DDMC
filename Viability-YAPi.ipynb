# YAP INHIBITORS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from msresist.figures.figure1 import TimePointFoldChange
from msresist.figures.figure4 import transform_YAPviability_data, plot_YAPinhibitorTimeLapse
```

```python
pd.set_option('display.max_colwidth', 1000)
pd.set_option('display.max_rows', 2000)
pd.set_option('display.max_columns', 1000)
```

```python
sns.set(style="whitegrid", font_scale=1.1, color_codes=True, palette="colorblind", rc={"grid.linestyle": "dotted", "axes.linewidth": 0.6})
```

## Untreated

```python
untreated = pd.read_csv("msresist/data/Validations/0nM.csv")
ut = TimePointFoldChange(untreated, 24)

fig, ax = plt.subplots(2, 4, figsize=(15, 7))
plt.subplots_adjust(wspace=0.3, hspace=0.35)

ut = transform_YAPviability_data(ut)
plot_YAPinhibitorTimeLapse(ax, ut)
```

```python
imatinib = pd.read_csv("msresist/data/Validations/Imatinib.csv")
im = TimePointFoldChange(imatinib, 24)

fig, ax = plt.subplots(2, 4, figsize=(15, 7), sharey=True)
plt.subplots_adjust(wspace=0.3, hspace=0.35)

im = transform_YAPviability_data(im)
plot_YAPinhibitorTimeLapse(ax, im)
```
